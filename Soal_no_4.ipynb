```python
corpus = [
    'Various sectors of life have changed due to the COVID-19 pandemic.',
    'One of the sectors affected is the education sector.',
    'Education is an important sector that requires appropriate solutions and policies.',
    'One of them is by studying at home.',
    'Studying at home is a solution that is very suitable and relevant to the pandemic conditions.',
    'First, by studying at home the number of virus spread due to interactions between people can be reduced.',
    'The spread of the virus can be through the air we breathe together in the same room, through touch, and other factors if we interact frequently.',
    'Although this can be overcome by wearing a mask and washing hands, but surely sometimes we forget and ignore it.',
    'Students and teachers can interact through online learning media such as zoom, g-meet, and so on so that learning continues.',
    'By studying at home can reduce the high risk of contracting the virus.',
    'Then, studying at home requires students to be more active and creative in solving their own problems.',
    'In the latest curriculum, which plays the biggest role is the activeness of students and teachers just to guide them.',
    'With this condition, students will get used to being more independent and creative in teaching and learning activities so that interactions with friends and teachers can still occur through online media.',
    'In this way, students will gain learning knowledge and also the ability to face various other problems later.',
    'In addition, studying at home can improve and enhance family relationships.',
    'With a lot of time spent at home, it means that a lot of time is also spent with loved ones.',
    'Children will ask their parents for help in learning and parents will help them.',
    'This can lead to continuous interaction so that mutual understanding can increase.',
    'That way, a sense of togetherness will grow and develop over time.',
    'Parents can also supervise their children more so that they know more about what their children have and what weaknesses they have.',
    'It can improve the ability of students who are learning.',
    'Studying at home is one of the right methods during this pandemic to apply.',
    'However, parental supervision becomes an important role to replace the teachers role.',
    'For this reason, it is necessary to have cooperation from various related parties for the progress of Indonesias education today.'
]

corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
vectorizer_X = vectorizer.fit_transform(corpus).todense()
vectorizer_X
```

```python
vectorizer.get_feature_names()
```

```python
from sklearn.metrics.pairwise import euclidean_distances

for i in range(len(vectorizer_X)):
    for j in range(i, len(vectorizer_X)):
        if i==j:
            continue
        jarak = euclidean_distances(vectorizer_X[i], vectorizer_X[j])
        print(f'Jarak dokumen {i+1} dan {j+1}: {jarak}')
```
